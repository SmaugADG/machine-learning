### 完善原始折线图1(辅助显示层)

```python
# 需求：画出某城市11点到12点每分钟温度变化图,温度范围15-18
import random
import matplotlib.pyplot as plt
# 可加可不加，依照环境添加
%matplotlib inline 
plt.figure()
# 1.准备数据
x=range(60)
y_shanghai=[random.uniform(15,18)for i in x]

# 2.创建画布
plt.figure(figsize=(20,8),dpi=80)

# 3.绘制图像
plt.plot(x,y_shanghai)
# 显示刻度：X轴：plt.xticks(x,**kwargs)
x_label=["11点{}分".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0,40,5))

# 添加网格显示
# 默认为TRUE，alpha：透明度
plt.grid(True,linestyle='--',alpha=0.5)

# 添加描述信息
# 添加x轴信息
plt.xlabel("时间")
# 添加Y轴信息
plt.ylabel("温度")
# 添加标题
plt.title("中午11点到12点温度变化图")


# 4.显示图像
plt.show()
```

### 完善原始折线图2(图像层)

```python
# 需求：再添加一个城市的温度变化
# 收集到北京当天温度变化情况，温度在1度到3度
# 需求：画出某城市11点到12点每分钟温度变化图,温度范围15-18
import random
import matplotlib.pyplot as plt
# 可加可不加，依照环境添加
%matplotlib inline 
plt.figure()
# 1.准备数据
x=range(60)
y_shanghai=[random.uniform(15,18)for i in x]
# 准备北京的数据
y_beijing=[random.uniform(1,2) for i in x]

# 2.创建画布
plt.figure(figsize=(20,8),dpi=80)

# 3.绘制图像
plt.plot(x,y_shanghai,color='r',linestyle='--',label="上海")
plt.plot(x,y_beijing,color='b',label="北京")
# 显示刻度：X轴：plt.xticks(x,**kwargs)
x_label=["11点{}分".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0,40,5))

# 显示图例
plt.legend()

# 添加网格显示
# 默认为TRUE，alpha：透明度
plt.grid(True,linestyle='--',alpha=0.5)

# 添加描述信息
# 添加x轴信息
plt.xlabel("时间")
# 添加Y轴信息
plt.ylabel("温度")
# 添加标题
plt.title("中午11点到12点温度变化图")


# 4.显示图像
plt.show()
```

### 多个坐标系显示-plt.subplots(面向对象画图)
    -matplotlib.pyplotlib.subplots(nrows=1,ncols=1,**fig_kw)

```python
# 需求：再添加一个城市的温度变化
# 收集到北京当天温度变化情况，温度在1度到3度
# 需求：画出某城市11点到12点每分钟温度变化图,温度范围15-18
import random
import matplotlib.pyplot as plt
# 可加可不加，依照环境添加
%matplotlib inline 
plt.figure()
# 1.准备数据
x=range(60)
y_shanghai=[random.uniform(15,18)for i in x]
# 准备北京的数据
y_beijing=[random.uniform(1,2) for i in x]


# 2.创建画布
# plt.figure(figsize=(20,8),dpi=80)
figure,axes=plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=80)


# 3.绘制图像
axes[0].plot(x,y_shanghai,color='r',linestyle='--',label="上海")
axes[1].plot(x,y_beijing,color='b',label="北京")
# 显示图例
axes[0].legend()
axes[1].legend()


# 显示刻度：X轴：plt.xticks(x,**kwargs)
x_label=["11点{}分".format(i) for i in x]
axes[0].set_xticks(x[::10])
axes[0].set_xticklabels(x_label)
axes[0].set_yticks(range(0,40,5))

axes[1].set_xticks(x[::10])
axes[0].set_xticklabels(x_label)
axes[1].set_yticks(range(0,40,5))



# 添加网格显示
# 默认为TRUE，alpha：透明度
axes[0].grid(True,linestyle='--',alpha=0.5)
axes[1].grid(True,linestyle='--',alpha=0.5)

# 添加描述信息
# 添加x轴信息
axes[0].set_xlabel("时间")
axes[1].set_xlabel("时间")
# 添加Y轴信息
axes[0].set_ylabel("温度")
axes[1].set_ylabel("温度")
# 添加标题
axes[0].set_title("上海中午11点到12点温度变化图")
axes[1].set_title("北京中午11点到12点温度变化图")

# 4.显示图像
plt.show()
```
